```python
## Imports and pandas settings

import pandas as pd
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

import requests
```

```python
## Constants
API_ENDPOINT = "https://fangraphs.com/api/players/{api_name}?playerid={player_id}&position={position}&type=0{extra}"

## Helper Functions

# Get dtypes for each columns of the df
def get_col_dtypes(df): 
    return df.dtypes.to_dict()

# Check the column type to see if it matches the expected type
def check_col_type(df, col_name, expected_type):
    dtypes = get_col_dtypes(df)
    return dtypes[col_name] == expected_type

# For all string/object columns in the df, remove all HTML tags
# This is a problem when pulling from FanGraphs data, as their JSON responses are messy
def remove_html_tags(df):
    for col in df.columns:
        if check_col_type(df, col, "object"):
            df[col] = df[col].str.replace(r'<[^<>]*>', '', regex=True)
    return df

def get_data(api_name, entry_point, player_id, position):
    r = requests.get(API_ENDPOINT.format(api_name = api_name, player_id = player_id, position = position, extra = ""))
    df = pd.json_normalize(r.json()[entry_point])
    df = remove_html_tags(df)
    return df


def get_stats(player_id, position):
    r = requests.get(API_ENDPOINT.format(api_name = "stats", player_id = player_id, position = position, extra = ""))
    df = pd.json_normalize(r.json()["data"])
    df = remove_html_tags(df)
    return df

def get_splits(player_id, position, season):
    r = requests.get(API_ENDPOINT.format(api_name = "splits", player_id = player_id, position = position, extra = "&season={}&split=".format(season)))
    df = pd.DataFrame.from_dict(r.json())
    df = remove_html_tags(df)
    return df
    
```

```python
df = get_splits("14162", "SS", "2022")
display(df)
```

```python
df = get_stats("14162", "SS")
display(df)
```

```python
df = get_data("stats", "data", "14162", "SS")
display(df)
```
